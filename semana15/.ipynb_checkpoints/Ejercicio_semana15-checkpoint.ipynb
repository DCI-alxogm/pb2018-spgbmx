```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sc
#Para activar los graficos en el notebook
%matplotlib inline
```

```python
f= np.loadtxt('IRCF.txt').T
np.shape(f)
```

```python
x=f[0] #frecuencia
y=f[1] #espectro
z=f[3]#*1e3 #incertidumbre
```

```python
plt.plot(x,y)
plt.errorbar(x,y,yerr=z)
plt.grid()
```

```python
plt.plot(z)
```

```python
#utilizando la libreria de constantes
h=sc.Planck
k=sc.Boltzmann
c=sc.c
```

```python
print(h,k,c)
```

```python
v=x #valor provisional,frecuencia
T=12 #valor provisional
a=(h*v**3)/c**2
b=(np.exp(h*v/k*T))
c=(1/b-1)
```

```python
def I(v,T):
    d=a*c
    return d
```

```python
plt.plot(I(v,T))
```
