```python
import numpy as np
import matplotlib.pyplot as plt
#Para activar los graficos en el notebook
%matplotlib inline     
```

```python
x=np.linspace(-10,10,100)
x2=np.linspace(0,5,5)
y=x**2*np.cos(x)-3*x**3 #funciones trigonometricas siempre dan valores en radianes
y2=x2**2*np.cos(x2)-3*x2**3
x
```

```python
#es mejor separar las celdas para hacer las graficas.
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
#Opción 1 para definir funciones
fx=lambda x: x**2*np.cos(x)-3*x**3
```

```python
fx(2)
```

```python
fx(x2)
```

```python
#opción 2 para definir funciones
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
fx2
```

```python
plt.plot(x,fx2(x,y))
```

```python
x3=np.random.rand(10)
fx2(x3,2)
```

```python
def recta(a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))
plt.plot(x,recta(x,3,-2))
plt.plot(x,recta(x,-3))
#para hacer cuadricula en la grafica
plt.grid()
```

```python
def recta(a,m=1,b=0,label='x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    
    return 
```

```python
x=np.linspace(-2,2,20)
recta(x)
recta(x,-2,1,'-x+1')
plt.legend()
plt.grid()
```
